# DATA DECOMPOSITION

```python
#Importing all the packages 
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
import pandas as pd
from dateutil.parser import parse 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#importing our dataset
df = pd.read_excel(r'/home/pragya/fyp/date_data.xlsx', parse_dates=['date'], index_col='date')
```

## Decomposition for SO2

```python
# Multiplicative Decomposition 
result_mul_so2 = seasonal_decompose(df['so2'], model='multiplicative', freq = 30)

# Additive Decomposition
result_add_so2 = seasonal_decompose(df['so2'], model='additive', freq = 30)

plt.rcParams.update({'figure.figsize': (10,10)})
result_mul_so2.plot().suptitle('Multiplicative Decompose', fontsize=22)
result_add_so2.plot().suptitle('Additive Decompose', fontsize=22)
plt.show()
```
