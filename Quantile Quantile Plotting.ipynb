# Quantile Quantile Plotting

```python
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
df = pd.read_excel(r'/home/pragya/fyp/date_data.xlsx')
```

```python
no2 = df['no2']
so2 = df['so2']
rspm = df['rspm']
spm = df['spm']
```

## QQ plotting for NO2

```python
fig = sm.qqplot(no2)
plt.show()
```

```python
#after fitting
fig = sm.qqplot(no2, fit=True, line='45')
plt.show()
```

## QQ plotting for SO2

```python
fig = sm.qqplot(so2)
plt.show()
```
