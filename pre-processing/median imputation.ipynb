```python
import pandas as pd
df = pd.read_excel (r'/home/pragya/fyp/abc.xlsx')
```

```python
df.columns = ['year', 'month', 'date', 'so2', 'no2','rspm','spm']
```

```python
so2_median = df.so2.median()
no2_median = df.no2.median()
rspm_median = df.rspm.median()
spm_median = df.spm.median()
```

```python
df.so2.fillna(so2_median, inplace = True) 
df.no2.fillna(no2_median, inplace = True) 
df.rspm.fillna(rspm_median, inplace = True) 
df.spm.fillna(spm_median, inplace = True) 
```

```python
df.to_excel(r'/home/pragya/fyp/abc_median.xlsx')
```
