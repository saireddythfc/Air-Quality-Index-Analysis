```python
import pandas as pd
df = pd.read_excel (r'/home/pragya/fyp/abc.xlsx')
```

```python
df.columns = ['year', 'month', 'date', 'so2', 'no2','rspm','spm']
```

```python
so2_mean = df.so2.mean()
no2_mean = df.no2.mean()
rspm_mean = df.rspm.mean()
spm_mean = df.spm.mean()
```

```python
df.so2.fillna(so2_mean, inplace = True) 
df.no2.fillna(no2_mean, inplace = True) 
df.rspm.fillna(rspm_mean, inplace = True) 
df.spm.fillna(spm_mean, inplace = True) 
```

```python
df.to_excel(r'/home/pragya/fyp/abc_mean.xlsx')
```
